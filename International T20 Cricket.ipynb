# 3. Print the top five teams by their win percentages.

```python
max_winners=df['info.outcome.winner'].value_counts()
```

```python
def evaluate(string):
    return eval(string)
```

```python
df['info.teams']=df['info.teams'].apply(evaluate)
```

```python
team1=[]
team2=[]
for x in df['info.teams']:
    team1.append(x[0])
for x in df['info.teams']:
    team2.append(x[1])
```

```python
total_matches_played=pd.Series(team1+team2)
```

```python
total_matches_played=total_matches_played.value_counts()
```

```python
total_matches_played
```

```python
max_winners=pd.DataFrame(data=max_winners).reset_index().rename(columns={'index':'Team','info.outcome.winner':'Total_wins'})
```

```python
max_winners
```

```python
total_matches_played=pd.DataFrame(data=total_matches_played).reset_index().rename(columns={'index':'Team',0:'Total_matches_played'})
```

```python
total_matches_played
```

```python
final_table=pd.merge(total_matches_played,max_winners,on='Team')
```

```python
final_table['Win_percentage']=round((final_table['Total_wins']/final_table['Total_matches_played'])*100,2)
```

```python
print('Top 5 teams with highest win percentage are:')
final_table.sort_values(by='Win_percentage',ascending=False).iloc[0:5]
```

# 4. Write a function to get the scorecard of each match. This function would take the innings value as argument and return two scorecard dataframes each for one team as shown below. So the first dataframe would contain the top 4 scorers of the team who batted first and the top 4 bowlers of the opponent team. And the second dataframe would contain the top 4 scorers of the team who batted second and the top 4 bowlers of the opponent team.

```python
df['innings']=df['innings'].apply(evaluate)
```

```python
df['innings'].loc[0][0]
```

```python
no1_inning_batsmen=[]
no1_inning_bowler=[]
for num in range(len(df['innings'].loc[0][0]['1st innings']['deliveries'])):
    for x,y in df['innings'].loc[0][0]['1st innings']['deliveries'][num].items():
        no1_inning_batsmen.append((y['batsman'],y['runs']['total']))
        if 'wicket' in y:
            no1_inning_bowler.append(y['bowler'])
no1_top4_batsmen=pd.DataFrame(no1_inning_batsmen).rename(columns={0:'Batsmen',1:'Runs'}).groupby('Batsmen').sum().sort_values(by='Runs',ascending=False).iloc[0:4].reset_index()
no1_top4_bowler=pd.DataFrame(pd.DataFrame(no1_inning_bowler).value_counts()).rename(columns={0:'wickets'}).reset_index().rename(columns={0:'Bowler'}).sort_values(by='wickets',ascending=False).iloc[0:4]
        
```

```python
no2_inning_batsmen=[]
no2_inning_bowler=[]
for num in range(len(df['innings'].loc[0][1]['2nd innings']['deliveries'])):
    for x,y in df['innings'].loc[0][1]['2nd innings']['deliveries'][num].items():
        no2_inning_batsmen.append((y['batsman'],y['runs']['total']))
        if 'wicket' in y:
            no2_inning_bowler.append(y['bowler'])
no2_top4_batsmen=pd.DataFrame(no2_inning_batsmen).rename(columns={0:'Batsmen',1:'Runs'}).groupby('Batsmen').sum().sort_values(by='Runs',ascending=False).iloc[0:4].reset_index()
no2_top4_bowler=pd.DataFrame(pd.DataFrame(no2_inning_bowler).value_counts()).rename(columns={0:'wickets'}).reset_index().rename(columns={0:'Bowler'}).sort_values(by='wickets',ascending=False).iloc[0:4]
```

```python
def score_card(list1):
    no1_inning_batsmen=[]
    no1_inning_bowler=[]
    team1=[]
    team2=[]
    for num in range(len(list1[0]['1st innings']['deliveries'])):
        for x,y in list1[0]['1st innings']['deliveries'][num].items():
            no1_inning_batsmen.append((y['batsman'],y['runs']['total']))
            if 'wicket' in y:
                no1_inning_bowler.append(y['bowler'])
    for x,y in list1[0].items():
        team1.append(y['team'])
    for x,y in list1[1].items():
        team2.append(y['team'])
        
    no1_top4_batsmen=pd.DataFrame(no1_inning_batsmen).rename(columns={0:'Batsmen',1:'Runs'}).groupby('Batsmen').sum().sort_values(by='Runs',ascending=False).iloc[0:4].reset_index()
    no1_top4_bowler=pd.DataFrame(pd.DataFrame(no1_inning_bowler).value_counts()).rename(columns={0:'wickets'}).reset_index().rename(columns={0:'Bowler'}).sort_values(by='wickets',ascending=False).iloc[0:4]
    
    no2_inning_batsmen=[]
    no2_inning_bowler=[]
    for num in range(len(list1[1]['2nd innings']['deliveries'])):
        for x,y in list1[1]['2nd innings']['deliveries'][num].items():
            no2_inning_batsmen.append((y['batsman'],y['runs']['total']))
            if 'wicket' in y:
                no2_inning_bowler.append(y['bowler'])
    no2_top4_batsmen=pd.DataFrame(no2_inning_batsmen).rename(columns={0:'Batsmen',1:'Runs'}).groupby('Batsmen').sum().sort_values(by='Runs',ascending=False).iloc[0:4].reset_index()
    no2_top4_bowler=pd.DataFrame(pd.DataFrame(no2_inning_bowler).value_counts()).rename(columns={0:'wickets'}).reset_index().rename(columns={0:'Bowler'}).sort_values(by='wickets',ascending=False).iloc[0:4]
    print('1st innings\n','\n','Team:',team1[0],'\n','\n',no1_top4_batsmen,'\n','\n','Team:',team2[0],'\n','\n',no1_top4_bowler)
    print('\n')
    print('2nd innings\n','\n','Team:',team2[0],'\n','\n',no2_top4_batsmen,'\n','\n','Team:',team1[0],'\n','\n',no2_top4_bowler)
```
