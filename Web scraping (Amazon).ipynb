```python
#Using requests and scrapy library to access the website and scrape the data on the first page

url='https://www.amazon.in/s?k=mobiles+5g&crid=2T9LPN1XBSU8W&sprefix=mobiles%2Caps%2C185&ref=nb_sb_ss_ts-doa-p_8_7'
header=response.json()['result'][3]
r=requests.get(url,headers=header)
response=TextResponse(r.url,body=r.text,encoding="utf-8")
```

```python
#Creating empty list in order to feed the scraped data into their respective list
title=[]
price=[]
rating=[]
# Using for loops to access the particular class from which we'll scaped the title,price and rating
for x in response.css('div.sg-row')[3:]: 
    if x.css('img.s-image')==[]:
        continue
    elif x.css('img.s-image')=='None':
        continue
    else:
        #Gonna append the output in their respective list which we'll use to create a dataframe
        
        title.append(x.css('img.s-image').xpath('@alt').get())
        price.append(x.css('span.a-price-whole::text').get())
        rating.append(f"{x.css('span.a-icon-alt::text').get().split()[0]}/5")
        
#Using Pandas library to create a dataframe with the help of list in which appended the data above

df=pd.DataFrame(title).rename(columns={0:'Title'})
df['Price']=price
df['Rating']=rating
df
```
