```python
#Importing necessary libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import scrapy
from scrapy.http import TextResponse
```

# Youtube Trending DataFrame (Movies)

```python
title=[]
channel_=[]
views=[]
uploaded_=[]

url='https://www.youtube.com/feed/trending?bp=4gIKGgh0cmFpbGVycw%3D%3D'

browser=webdriver.Chrome()
browser.get(url)
browser.maximize_window()
response=TextResponse(browser.current_url,body=browser.page_source,encoding='utf-8')
overall=browser.find_elements(By.CSS_SELECTOR,"a#video-title")

for x in overall:
    time.sleep(3)
    title.append(x.get_attribute('title'))
    a=x.get_attribute('aria-label')
    channel_.append(' '.join(a.split(' by')[1].split('ago')[0].split()[0:-2]))
    views.append(a.split('views')[0].split()[-1])
    uploaded_.append(" ".join(a.split('ago')[0].split()[-2::]))
browser.quit()

df=pd.DataFrame(title).rename(columns={0:'Title'})
df['Channel']=channel_
df['Views']=views
df['Uploaded']=uploaded_
df
```
