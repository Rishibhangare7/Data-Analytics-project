```python
from bs4 import BeautifulSoup
import lxml
import requests
```

# Scraping Book's Title, Rating and Price in 1st page

```python
# Setting the url of 1st page and then requesting to get that url and then converting that into a soup using BeautifulSoup

url="http://books.toscrape.com/catalogue/page-1.html"
res=requests.get(url)
soup=BeautifulSoup(res.content,'lxml')
```

```python
# Creating a dictionary in order to convert rating numbers which are in words into numbers by putting them as value

dict1={'One':1,'Two':2,'Three':3,'Four':4,'Five':5}
```

```python
# Using BeautifulSoup methods in order to grab the title,rating and price of the books in the 1st page.

for x in soup.select('.product_pod'):
    print(f"{x.contents[1].find('img')['alt']}\nRating-{dict1[x.contents[3]['class'][1]]}/5\nPrice- {x.select('.price_color')[0].get_text()}\n")
```

# Scraping book titles from all pages with a rating of 5 out of 5

```python
# Using same BeautifulSoup methods for grabbing the title and price of the book with a rating of 5 from all the pages.
# With the help of While loops we can scan through every page and grab whatever we need.

print('Book title and price with a rating of 5 out of 5')
print('\n')
page_still_valid=True
page=1
while page_still_valid:
    url=f'http://books.toscrape.com/catalogue/page-{page}.html'
    res=requests.get(url)
    if 'Not Found' in res.text:
        break
    soup=BeautifulSoup(res.content,'lxml')
    for x in soup.select('.product_pod'):
        if x.contents[3]['class'][1]=='Five':
            print(f"{x.contents[1].find('img')['alt']}\nPrice-{x.select('.price_color')[0].get_text()}\n")
    page=page+1
```

# Creating a DataFrame with Book's title,rating and price from all pages

```python
# importing necessary libraries in order to create a dataframe
import pandas as pd
import numpy as np
```

```python
dict1={'One':1,'Two':2,'Three':3,'Four':4,'Five':5}
```

```python
# Using same BeautifulSoup methods for grabbing the title,rating and price of the books from all pages.
# With the help of While loops we can scan through every page and grab whatever we need.
# After grabbing them appending them to their respective list which we'll need in order to create a dataframe columns

page_still_valid=True
Books_title=[]
Rating=[]
Price=[]
page=1
while page_still_valid:
    url=f'http://books.toscrape.com/catalogue/page-{page}.html'
    res=requests.get(url)
    if 'Not Found' in res.text:
        break
    soup=BeautifulSoup(res.content,'lxml')
    for x in soup.select('.product_pod'):
        Books_title.append(x.contents[1].find('img')['alt'])
        Price.append(x.select('.price_color')[0].get_text())
        Rating.append(str(dict1[x.contents[3]['class'][1]])+'/5')
    page=page+1
    
```

```python
# Creating a dataframe using pandas and creating columns with the help of the lists which we created before

Books_df=pd.DataFrame(Books_title).rename(columns={0:'Title'})
Books_df['Rating']=Rating
Books_df['Price']=Price
```
